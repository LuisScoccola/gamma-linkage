```python
import sys
sys.path.append("../")

from persistable import Persistable, PersistableInteractive

import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl
%load_ext autoreload
%autoreload 2
```

```python
# Download dataset to a variable and display it
# We use an example dataset available at the accelerated HDBSCAN repository
# [https://github.com/scikit-learn-contrib/hdbscan] (McInnes, Healy, Astels).

from urllib.request import urlopen
from io import BytesIO

url = "https://github.com/scikit-learn-contrib/hdbscan/blob/4052692af994610adc9f72486a47c905dd527c94/notebooks/clusterable_data.npy?raw=true"
f = urlopen(url) ; rf = f.read() ; X = np.load(BytesIO(rf))
#plt.figure(figsize=(3,3)) ; _ = plt.scatter(X.T[0], X.T[1], s=1, alpha = 0.5)
```

```python
p = Persistable(X)
pi = PersistableInteractive(p)
#pi.parameter_selection()
```

```python
cl = pi.cluster() #extend_clustering_by_hill_climbing=True,n_neighbors_extend_cluster=10, n_iterations_extend_cluster=4)
```

```python
plt.figure(figsize=(3,3))
_ = plt.scatter(X[cl==-1,0], X[cl==-1,1], s=2, alpha = 0.2, c = "grey")
_ = plt.scatter(X[cl!=-1,0], X[cl!=-1,1], s=2, alpha = 0.5, c = cl[cl!=-1], cmap="Accent")
```
