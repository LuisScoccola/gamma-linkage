## Construct dataset and display it

```python
from sklearn import datasets

N_DATAPOINTS = 1000

n_true_points = int(N_DATAPOINTS * 0.7) ; X, _ = datasets.make_moons(n_samples=n_true_points, noise=0.1, random_state=0)
np.random.seed(0) ; n_noise = N_DATAPOINTS - n_true_points ; noise = (np.random.random_sample((n_noise,2)) - 0.4) * 4 ; X = np.vstack((X,noise))
fig1, ax1 = plt.subplots() ; fig1.set_figheight(4) ; fig1.set_figwidth(4)
_ = plt.scatter(X.T[0], X.T[1], s=5, alpha = 0.5)
```

## Run gamma-linkage

### 1. Fit metric probability space

If one wants to use a different kernel, or knows one will not use curves
with a k-intercept that is too large, one can, for instance, fit as follows
[x.fit(kernel="epanechnikov", maxk=300)]

### 2. Compute the hierarchical clustering using the curve gamma

See below for how we have chosen the parameters for the curve gamma.

### 3. Compute the persistence-based flattening and display it

See below for how we have chosen the parameter for the persistence-based flattening.


```python
# 1.
x = MPSpace(X)
x.fit()

# 2.
K_INTERCEPT = 0.1
S_INTERCEPT = 0.33
gamma = Gamma_curve.linear_interpolator_alpha_s_indexed(K_INTERCEPT, S_INTERCEPT)
hc = x.gamma_linkage(gamma)

# 3.
N_CLUSTERS = 2
c = hc.persistence_based_flattening(num_clusters=N_CLUSTERS)
```

```python
fig1, ax1 = plt.subplots() ; fig1.set_figheight(4) ; fig1.set_figwidth(4)
_ = plt.scatter(X.T[0], X.T[1], s=5, alpha = 0.5, c = -c[1], cmap="Accent")
```

## Parameter selection

### 1. Run gamma-linkage with a range of curves

We compute S_INTERCEPT = max_s to be the smallest radius for which the Rips graph of the dataset becomes connected.
We choose a range of k from 0.01 to 0.2 (1/100 to 1/5 of the dataset).
We run gamma-linkage 500 times with parameters k and max_s.
We display the total persistences as the parameters change.

#### Note: the most persistent point (which has infinite total persistence) is not displayed.
#### Note: this cell runs gamma-linkage 500 times, so it can take several seconds to run

The plot suggests that, across paramaters k, there are 2 high persistent classes: the curve that is separated from the rest, and the infinite persistence one that is not displayed.
We choose K_INTERCEPT = 0.1 and S_INTERCEPT = 0.33.

### 2. Display the persistence diagram of gamma-linkage with the chosen curve

We do not display the infinite persistence point.

We see that there is 1 well-separated point from the rest.
This confirms that there are 2 high persistence points.

We choose N_CLUSTERS = 2 for the persistence-based flattening.

```python
# 1.
max_s = x.connection_radius()
initial_k = 0.01
final_k = 0.2
N_PARAMETERS = 500

TOL = 1e-15 ; INF = 1e15

def total_persistences(bd : np.array) -> np.array :
    pers = np.abs(bd[:,0] - bd[:,1])
    pers = pers[pers > TOL]
    pers = pers[pers < INF]
    return pers

gammas = [ Gamma_curve.linear_interpolator_alpha_s_indexed(k, max_s) for k in np.linspace(initial_k, final_k, num=N_PARAMETERS) ]
# using end="infinty" when constructing the persistence diagram makes the most persistent point die at infinity, and thus be ignored when constructing the total persistence curves, below
persistences = [ total_persistences(x.gamma_linkage(curve).PD(end="infinity")[0]) for curve in gammas ]
```

```python
points = []
for i in range(N_PARAMETERS) :
    for y in persistences[i] :
        points.append([i,y])
points = np.array(points)

fig1, ax1 = plt.subplots() ; fig1.set_figheight(5) ; fig1.set_figwidth(10) ; plt.xlabel("k intercept") ; plt.ylabel("total persistences") ; fig1.set_figheight(5) ; fig1.set_figwidth(8)

_ = plt.xticks(np.linspace(0, N_PARAMETERS, num=10), np.rint(100*np.linspace(initial_k, final_k, num=10))/100 )
_ = plt.scatter(points.T[0], points.T[1], s = 1, alpha = 1)
```

```python
# 2.

pd = hc.PD(end="infinity")

fig1, ax1 = plt.subplots() ;  fig1.set_figheight(4) ; fig1.set_figwidth(4)
plt.scatter(pd[0].T[0], pd[0].T[1], s = 10) ; ax1.axline((0, 0), slope=1, color='grey')
```

## What if we use a horizontal curve (like HDBSCAN)?

We now use the same range for k, but we use horizontal curves, so we do not expect the diagram of total persistences to consist of continuous curves.

```python
# 1.
gammas_h = [ Gamma_curve.constant_k_alpha_s_indexed(k) for k in np.linspace(initial_k, final_k, num=N_PARAMETERS) ]
persistences_h = [ total_persistences(x.gamma_linkage(curve).PD(end="infinity")[0]) for curve in gammas_h ]
```

```python
points_h = []
for i in range(N_PARAMETERS) :
    for y in persistences_h[i] :
        points_h.append([i,y])
points_h = np.array(points_h)

fig1, ax1 = plt.subplots() ; fig1.set_figheight(5) ; fig1.set_figwidth(10) ; plt.xlabel("k intercept") ; plt.ylabel("total persistences") ; fig1.set_figheight(5) ; fig1.set_figwidth(8)


_ = plt.xticks(np.linspace(0, N_PARAMETERS, num=10), np.rint(100*np.linspace(initial_k, final_k, num=10))/100 )
_ = plt.scatter(points_h.T[0], points_h.T[1], s = 1, alpha = 1)
```
