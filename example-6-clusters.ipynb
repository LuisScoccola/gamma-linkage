## Download dataset to a variable and display it

```python
from urllib.request import urlopen
from io import BytesIO

url = "https://github.com/scikit-learn-contrib/hdbscan/blob/4052692af994610adc9f72486a47c905dd527c94/notebooks/clusterable_data.npy?raw=true"

f = urlopen(url)
rf = f.read()
X = np.load(BytesIO(rf))

fig1, ax1 = plt.subplots() ; fig1.set_figheight(5) ; fig1.set_figwidth(8) ; _ = plt.scatter(X.T[0], X.T[1], s=10, alpha = 0.5)
```

## Parameter selection

### 1. Run gamma-linkage with a range of curves

We compute S_INTERCEPT = max_s to be the smallest radius for which the Rips graph of the dataset becomes connected.
We choose a range of k from 0.005 to 0.1 (1/200 to 1/10 points of the dataset).
We run gamma-linkage 500 times with parameters k and max_s.
We display the total persistences as the parameters change.

#### Note: the most persistent point (which has infinite total persistence) is not displayed.
#### Note: this cell runs gamma-linkage 500 times, so it can take several seconds to run. Improved runtime will be addressed in the next version. 

The plot suggests that, across paramaters k, there are 6 high persistent classes: the 5 curves that are separated from the rest, and the infinite persistence one that is not displayed.

We choose K_INTERCEPT = 0.03 and S_INTERCEPT = 0.1.

### 2. Display the persistence diagram of gamma-linkage with the chosen curve

We do not display the infinite persistence point.

We see that there are 5 well-separated points from the rest.
This confirms that there are 6 high persistence points.

We choose N_CLUSTERS = 6 for the persistence-based flattening.

```python
# 1.
max_s = x.connection_radius()
initial_k = 0.005
final_k = 0.1
N_PARAMETERS = 500

TOL = 1e-15 ; INF = 1e15

def total_persistences(bd : np.array) -> np.array :
    pers = np.abs(bd[:,0] - bd[:,1])
    pers = pers[pers > TOL]
    pers = pers[pers < INF]
    return pers

gammas = [ Gamma_curve.linear_interpolator_alpha_s_indexed(k, max_s) for k in np.linspace(initial_k, final_k, num=N_PARAMETERS) ]
# using end="infinty" when constructing the persistence diagram makes the most persistent point die at infinity, and thus be ignored when constructing the total persistence curves and the persistence diagram, below
persistences = [ total_persistences(x.gamma_linkage(curve).PD(end="infinity")[0]) for curve in gammas ]
```

```python
points = []
for i in range(N_PARAMETERS) :
    for y in persistences[i] :
        points.append([i,y])
points = np.array(points)

fig1, ax1 = plt.subplots() ; fig1.set_figheight(5) ; fig1.set_figwidth(10) ; plt.xlabel("k intercept") ; plt.ylabel("total persistences") 

_ = plt.xticks(np.linspace(0, N_PARAMETERS, num=10), np.rint(100*np.linspace(initial_k, final_k, num=10))/100 )
_ = plt.scatter(points.T[0], points.T[1], s = 1, alpha = 1)
```

```python
# 2.

pd = hc.PD(end="infinity")

fig1, ax1 = plt.subplots() ;  fig1.set_figheight(4) ; fig1.set_figwidth(4)
plt.scatter(pd[0].T[0], pd[0].T[1], s = 10) ; ax1.axline((0, 0), slope=1, color='grey')
```

## Running "skew" curves again, but much fewer times

Because of stability of "skew" lines, we can run the algorithm fewer times and construct a 100%-confidence region for the total persistences.
This is sufficient to detect good separation.
We do this with 50 parameters (instead of 500).


```python
N_PARAMETERS_ = 50

gammas_ = [ Gamma_curve.linear_interpolator_alpha_s_indexed(k, max_s) for k in np.linspace(initial_k, final_k, num=N_PARAMETERS_) ]
persistences_ = [ total_persistences(x.gamma_linkage(curve).PD(end="infinity")[0]) for curve in gammas_ ]
```

```python
params = np.linspace(initial_k, final_k, num=N_PARAMETERS_)

bounds = [np.abs(params[i] - params[i+1]) * max_s * max(1/params[i], 1/params[i+1]) for i in range(N_PARAMETERS_ - 1)]

fig1, ax1 = plt.subplots() ; fig1.set_figheight(5) ; fig1.set_figwidth(10) ; plt.xlabel("k intercept") ; plt.ylabel("total persistences") ; 

for i in range(N_PARAMETERS_-1) :
    for y in persistences_[i] :
        possibles = [ z for z in persistences_[i+1] if np.abs(y-z) <= bounds[i]]
        for z in possibles :
            t = 0.5 + (z-y)/(2*bounds[i])
            a = (i + t, y + t * bounds[i])
            b = (i + (1-t), y - (1-t) * bounds[i])
            rectangle = plt.Polygon([(i,y), a, (i+1,z), b], linewidth=0.5, color="lightsteelblue", zorder = 1)
            _ = ax1.add_patch(rectangle)

ax1.set_rasterized(True)

points_ = []
for i in range(N_PARAMETERS_) :
    for y in persistences_[i] :
        points_.append([i,y])
points_ = np.array(points_)
        
_ = plt.scatter(points_.T[0], points_.T[1], s = 10, alpha = 1, zorder=2, rasterized=True)
_ = plt.xticks(np.linspace(0, N_PARAMETERS_, num=10), np.rint(100*np.linspace(initial_k, final_k, num=10))/100 )

_ = plt.ylim([0,0.035])
```
